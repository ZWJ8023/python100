```python
"""
练习1:
    在屏幕上显示跑马灯文字
    ------------------
    

"""
import os
import time

def main():
    input_='北京欢迎你.......'
    while :
        os.system('clear') #清理屏幕上的输出
        print(input_)
        time.sleep(0.1) #间隔1秒
        input_=input_[1:] + input_[0] #利用字符串的切片功能
        
main()
```

```python
"""
练习2.
    设计一个函数产生指定长度的验证码，验证码由大小写字母和数字组成
    ---------------------------------------------------
    all_chars:将所有的数字字母定义成字符串
    因为字符串切片从0开始，所以去最后一个元素时，长度-1
    定义一个空的字符串，用for循环遍历出元素将其添加到空的字符串中

"""
import random


def main(yanzhengma): #若输入的是yanzhengma=4,则说明这个验证码的长度设为固定长度
    all_chars = '0123456789abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
    last_num = len(all_chars) -1
    new_chars=' '
    for i in range(yanzhengma):
        num=random.randint(0,last_num)
        new_chars +=  all_chars[num]
    print(new_chars)
    
main(6) #改变main()中的数，则可得到不同长度的验证码
```

```python
"""
练习4.
    设计一个函数返回传入的列表中的最大和第二大的元素的值
---------------------------------------------
    首先判断列表的第0位和第1位的大小
    用for循环遍历其他位的元素
    如果遍历出的x[i]的值>num1,则往后推，num2 = num1

"""

def max2(x):
    num1,num2 = (x[0],x[1]) if  x[0]>x[1] else (x[1],x[0])
    for i in range(2,len(x)):
        if x[i] > num1:
            num2 = num1
            num1 = x[i]
        elif x[i] > num2:
            num2 = x[i]
    print(num1,num2)
    
max2('1683')


```
